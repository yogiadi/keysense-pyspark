# KeySense Quickstart (PySpark)

This minimal example creates a tiny DataFrame and runs `KeySense.evaluate()`.
It doesn't download any data — you can run it as-is if PySpark is available in your environment.

```python
# If PySpark is not installed, uncomment the next line to install it in notebook environments
# !pip install pyspark

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("KeySenseQuickstart").getOrCreate()
spark
```

```python
# Tiny demo dataset (4 rows)
data = [
    ("u1", "s1", "2025-08-01", "mobile"),
    ("u2", "s2", "2025-08-01", "desktop"),
    ("u1", "s1", "2025-08-02", "mobile"),
    ("u3", "s3", "2025-08-01", "mobile"),
]
cols = ["user_id", "session_id", "event_date", "device_id"]
df = spark.createDataFrame(data, cols)
df.show(truncate=False)
```

```python
from keysense import KeySense  # if you didn't update __init__.py, use: from keysense.profiler import KeySense

ks = KeySense(df, time_col="event_date", max_combo_len=3)
results = ks.evaluate()

# Show the top few results
for r in results[:5]:
    print(r)
```
